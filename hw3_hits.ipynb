```python
import numpy as np
import time

file_name = ['graph_1.txt','graph_2.txt','graph_3.txt','graph_4.txt','graph_5.txt','graph_6.txt','project1_data.txt','project1_data2.txt']
#file_name = ['graph_1.txt','graph_1_newh.txt','graph_2.txt','graph_2_newh.txt','graph_3.txt','graph_3_newh.txt']
#file_name = ['graph_1.txt','graph_1_newp.txt','graph_2.txt','graph_2_newp.txt','graph_3.txt','graph_3_newp.txt']
for file in file_name:
    start = time.time()
    max_value = 1
    graph_dict = {}
    tmp_list = []
    ha_old_dict = {}
    ha_new_dict = {}
    print('file name:',file)
    with open('./hw3dataset/' + file) as f:
        txt_data = f.read().split('\n')
    for item in txt_data:
        if item[:item.find(',')] != '' and item[item.find(',')+1:] != '':
            max_value = max(int(item[:item.find(',')]),int(item[item.find(',')+1:]),max_value)
    for node in range(1, max_value+1):
        ha_old_dict[str(node)] = [1,1] 
        ha_new_dict[str(node)] = [0,0] 
        graph_dict[str(node)] = []
    for item in txt_data:  
        if item[:item.find(',')] != '' and item[item.find(',')+1:] != '':
            source = item[:item.find(',')]
            target = item[item.find(',')+1:]
            if graph_dict[source] == []:
                graph_dict[source] = [target]
            else:
                tmp_list = graph_dict[source]
                tmp_list.append(target)
                graph_dict[source] = tmp_list
                tmp_list = []
    iteration = 0
    total = 1
    norm_dict = {}
    new_norm_dict = {}
    while total > 0.01:       
        iteration += 1
        for ite in ha_old_dict.keys():   
            authority = 0
            hubbiness = 0            
            # hunbness
            for hub in graph_dict[ite]:
                if hub in ha_old_dict:
                    hubbiness += ha_old_dict[hub][1]                
            #authority
            for va in ha_old_dict.keys():
                if ite in graph_dict[va]:
                    authority += ha_old_dict[va][0] 
            ha_new_dict[ite] = [hubbiness,authority]
        
        ha_old_dict = ha_new_dict.copy()
  
        tmp_ha_sum = 0
        tmp_au_sum = 0
        for norm in ha_new_dict.values():
            tmp_ha_sum += norm[0]
            tmp_au_sum += norm[1]        
            
        if iteration > 1:
            total = 0
            for norm in ha_new_dict.keys():
                new_norm_dict[norm] = [ha_new_dict[norm][0]/tmp_ha_sum, ha_new_dict[norm][1]/tmp_au_sum]
            for sub in ha_new_dict.keys():
                total += abs(new_norm_dict[sub][1] - norm_dict[sub][1]) 
                
            norm_dict = dict(new_norm_dict)
        else:
            for norm in ha_new_dict.keys():
                norm_dict[norm] = [ha_new_dict[norm][0]/tmp_ha_sum, ha_new_dict[norm][1]/tmp_au_sum]
    print('number of itrations:',iteration)
    end = time.time()  
    print('time cost:',end-start,'s')   
    for node in new_norm_dict.keys():
        with open('./HIST_result/' + file, 'a') as fi:
            hubb = ''
            hubb = 'hub '+ str(node)+ ' : ' + str(new_norm_dict[node][0]) + '\n'
            fi.write(hubb)
        print('hub ',str(node), ':' , new_norm_dict[node][0])
    for node in new_norm_dict.keys():
        with open('./HIST_result/' + file, 'a') as fi:
            auth = ''
            auth = 'authority '+str(node)+ ' : '+ str(new_norm_dict[node][1]) + '\n'
            fi.write(auth)
        print('authority',str(node), ':', new_norm_dict[node][1])
    print('-'*80)      
```
